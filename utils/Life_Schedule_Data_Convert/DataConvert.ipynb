```python
import csv
```

```python
SKD_HC = { '平日' : 'SKD_HC.W, ',
           '休日' : 'SKD_HC.H, ' }
SKD_V  = { '平日'   : 'SKD_V.W,  ',
           '休日在' : 'SKD_V.HH, ',
           '休日外' : 'SKD_V.HA, ' }
SKD_W  = { '平日（大）'     : 'SKD_W.W_L, ',
           '平日（中）'     : 'SKD_W.W_M, ',
           '平日（小）'     : 'SKD_W.W_S, ',
           '休日在宅（大）' : 'SKD_W.HH_L,',
           '休日在宅（小）' : 'SKD_W.HH_S,',
           '休日外出'       : 'SKD_W.HA,  ' }
SKD_L  = { '平日'   : 'SKD_L.W,  ',
           '休日在' : 'SKD_L.HH, ',
           '休日外' : 'SKD_L.HA, ' }
SKD_EA = { '平日'   : 'SKD_EA.W,  ',
           '休日在' : 'SKD_EA.HH, ',
           '休日外' : 'SKD_EA.HA, ' }
SKD_CK = { '平日'   : 'SKD_CK.W  ',
           '休日在' : 'SKD_CK.HH ',
           '休日外' : 'SKD_CK.HA ' }
```

```python
def dateconvert(s):
    
    def addSpace(n):
        if int(n) < 10:
            return ' '
        else:
            return ''
        
    d = s.split('月')
    month = d[0]
    day = d[1].replace('日','')
    return '\'' + month + '/' + day + '\'' + addSpace(month) + addSpace(day)
```

```python
print( dateconvert('1月1日') + '/')
print( dateconvert('1月2日') + '/')
print( dateconvert('1月10日') + '/')
print( dateconvert('12月1日') + '/')
print( dateconvert('12月2日') + '/')
print( dateconvert('12月10日') + '/')
```

```python
with open('住宅02-11-03_その他_生活スケジュール_生活スケジュール（データ）_131216.csv', 'r') as f:
    reader = csv.reader(f)
    header = next(reader)  # ヘッダーをスキップ

    for row in reader:
                
        print('    ' + dateconvert(row[0]) + ': SKD(' 
              + SKD_HC[row[1]]
              + SKD_V[row[2]]
              + SKD_W[row[3]]
              + SKD_L[row[4]]
              + SKD_EA[row[5]]
              + SKD_CK[row[6]]
              + '),' )
```

```python
with open('住宅02-11-03_その他_生活スケジュール_生活スケジュール（データ）_131216.csv', 'r') as f:
    reader = csv.reader(f)
    header = next(reader)  # ヘッダーをスキップ
    
    s = ""

    for row in reader:
        
        s = s + dateconvert(row[0]) + ','
    
    print ( '[' + s + ']')
```
